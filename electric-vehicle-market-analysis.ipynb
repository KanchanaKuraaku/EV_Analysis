```python
year = [ '2015', '2016' , '2017' , '2018' , '2019' , '2020' , '2021' , '2022' , '2023' , '2024']
total = ev_sales[year].sum()
graph = pd.DataFrame({'Year': total.index, 'Total Production': total.values})
print(graph.info())
print(graph)
```

```python
graph.plot( x = "Year" , y = "Total Production")
Total_Registered_vehicles = graph['Total Production'].sum()
print("Total Registered Vehicles - ",Total_Registered_vehicles)
Operational_charging_points = operational_pc['No. of Operational PCS'].sum()
print("Operational Charging Points -",Operational_charging_points)
```

```python
ev_sales['Total Production'] = ev_sales[year].sum(axis=1)
ev_sales.info()
```

```python
maker = ev_sales.groupby(['Maker','Cat'])['Total Production'].sum().reset_index()
maker['Total Production'] = maker['Total Production'].sort_values(ascending = False)
print(maker.head())
print(maker)
```

```python
two_wheeler_maker = maker.loc(maker['Cat'] == '2W')
```
